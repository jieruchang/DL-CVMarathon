```python
import cv2
import time
import numpy as np
import matplotlib.pyplot as plt

img_path = 'data/lena.png'
img = cv2.imread(img_path, cv2.IMREAD_COLOR)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

## 上下左右翻轉圖片

```python
# 水平翻轉 (horizontal)
img_hflip = img[:, ::-1, :]

# 垂直翻轉 (vertical)
img_vflip = img[::-1, :, :]

# 水平 + 垂直翻轉
img_hvflip = img[::-1, ::-1, :]

# 組合 + 顯示圖片
hflip = np.hstack((img, img_hflip))
vflip = np.hstack((img_vflip, img_hvflip))
img_flip = np.vstack((hflip, vflip))
plt.imshow(img_flip)

# while True:
#     cv2.imshow('flip image', img_flip)
#     k = cv2.waitKey(0)
#     if k == 27:
#         cv2.destroyAllWindows()
#         break
```

## 平移幾何轉換

```python
# 設定 translation transformation matrix
# x 平移 50 pixel; y 平移 100 pixel
M = np.array([[1, 0, 50],
             [0, 1, 100]], dtype = np.float32) # remember change type to float32
shift_img = cv2.warpAffine(img, M, (img.shape[1], img.shape[0]), 
                           borderValue=(0, 0, 0)) # image border show black # white: (255, 255, 255)

# 組合 + 顯示圖片
img_shift = np.hstack((img, shift_img))
plt.imshow(img_shift)
plt.show()

# while True:
#     cv2.imshow('shift image', img_shift)
#     k = cv2.waitKey(0)
#     if k == 27:
#         cv2.destroyAllWindows()
#         break
```
